```python
# Modified from Source: https://pythonhosted.org/sense-hat 

# Import dependencies
from sense_hat import SenseHat

# Initialise Sense HAT 
sense = SenseHat()
```

```python
# Display A Message on the Sense HAT LEDs
sense.show_message("Hello world!")
```

```python
# Display A Symbol on the Sense HAT LEDs
X = [255, 0, 0]  # Red
O = [255, 255, 255]  # White

question_mark = [
O, O, O, X, X, O, O, O,
O, O, X, O, O, X, O, O,
O, O, O, O, O, X, O, O,
O, O, O, O, X, O, O, O,
O, O, O, X, O, O, O, O,
O, O, O, X, O, O, O, O,
O, O, O, O, O, O, O, O,
O, O, O, X, O, O, O, O
]

sense.set_pixels(question_mark)
```

```python
# Set all the Sense HAT LEDs
sense.clear(0, 255, 0)
```

```python
# Set A Pixel using (x, y, r, g, b)
sense.set_pixel(0, 0, 255, 0, 0)
```

```python
# Clear the Sense HAT LEDs
sense.clear()
```

```python
# Get the Temperature
temp = sense.get_temperature()
print(f"Temperature: {temp:.2f} C")
sense.show_message(f"Temperature: {temp:.2f} C")
```

```python
# Get the Humidity
humidity = sense.get_humidity()
print(f"Humidity: {humidity:.2f} %")
sense.show_message(f"Humidity: {humidity:.2f} %")
```

```python
# Get the Pressure
pressure = sense.get_pressure()
print(f"Pressure: {pressure:.2f} Millibars")
sense.show_message(f"Pressure: {pressure:.2f} Millibars")
```
