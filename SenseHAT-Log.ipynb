```python
# Import dependencies
import time
import json
from datetime import datetime
from sense_hat import SenseHat

# Initialise Sense HAT 
sense = SenseHat()
```

```python
# Loop to take temperature reading and save to log
while True:
    temperature = sense.get_temperature()

    record = {
        "timestamp": datetime.now().isoformat(),
        "temperature": temperature
    }
    
    # Save the record to the log
    with open("temperature_log.json", "a") as f:
        json.dump(record, f)
        f.write("\n")
    
    # Output to console
    print(f"Saved {record}")
    
    # Pause for 5 Seconds
    time.sleep(5)
```
