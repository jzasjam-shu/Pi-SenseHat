### Wiring Instructions

**LED**
- Short Leg → GND rail/row on breadboard
- Long Leg → Same column as resistor

**Resistor**
- Resistor → Same column as LED long leg
- Resistor (other end) → Another column on the breadboard

**Jumper Wires (M/F)**
- Green (in the example, use any colour)
    - M → Same column as resistor
    - F → GPIO4 (Pin 7) on Raspberry Pi
- Black (in the example, use any colour)
    - M → GND rail/row on the breadboard 
    - F → GND (Pin 6)

**Pins Reference**

Hint: Use command `pinout` in the Raspberry Pi terminal to view the GPIO pin information

### Prerequisite: Enable I2C for GPIO Pins Access

1. On the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, then select ‘Preferences’ and ‘Raspberry Pi Configuration’

<img width="400" alt="image" src="../images/config.png" />

2. In the Raspberry Pi Configuration window select the ‘Interfaces’ tab, then toggle I2C so that it is enabled and click ‘OK’

<img width="400" alt="image" src="../images/i2c.png" />

3. You will be required to reboot the Pi if you have made changes


```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# TEST A SINGLE LED
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

# Modified from Source: https://gpiozero.readthedocs.io/en/stable/recipes.html#led 

from gpiozero import LED
from time import sleep

led = LED(4)        # LED with reference to the GPIO Pin

led.on()            # Turn on the LED
sleep(0.5)
led.off()           # Turn off the LED
sleep(0.5)

while True:
    
    led.toggle()    # Toggle the LED state (flash)
    sleep(0.5)
```
