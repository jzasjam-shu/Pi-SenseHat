## Light an LED from Code



### Wiring Instructions

**LED**
- Short Leg → GND rail/row on breadboard
- Long Leg → Same column as resistor

**Resistor**
- Resistor → Same column as LED long leg
- Resistor (other end) → Another column on the breadboard

**Jumper Wires (M/F)**
- Green (in the example, use any colour)
    - M → Same column as resistor
    - F → GPIO4 (Pin 7) on Raspberry Pi
- Black (in the example, use any colour)
    - M → GND rail/row on the breadboard 
    - F → GND (Pin 6)

```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# TEST A SINGLE LED
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

# Modified from Source: https://gpiozero.readthedocs.io/en/stable/recipes.html#led 

from gpiozero import LED
from time import sleep

led = LED(4)        # LED with reference to the GPIO Pin

led.on()            # Turn on the LED
sleep(0.5)
led.off()           # Turn off the LED
sleep(0.5)

while True:
    
    led.toggle()    # Toggle the LED state (flash)
    sleep(0.5)
```
