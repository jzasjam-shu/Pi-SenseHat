### Pins Reference
- **GND** → GND on Raspberry Pi
- **VIN** (VCC) → 3.3V or 5V on Raspberry Pi 
- **SCL** (Serial Clock Line) → GPIO3 (Pin 5) on Raspberry Pi
- **SDA** (Serial Data Line) → GPIO2 (Pin 3) on Raspberry Pi

**Pins Reference**

Hint: Use command `pinout` in the Raspberry Pi terminal to view the GPIO pin information

### Enable I2C for GPIO Pins Access

1. On the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, then select ‘Preferences’ and ‘Raspberry Pi Configuration’

<img width="400" alt="image" src="../images/config.png" />

2. In the Raspberry Pi Configuration window select the ‘Interfaces’ tab, then toggle I2C so that it is enabled and click ‘OK’

<img width="400" alt="image" src="../images/i2c.png" />

3. You may be required to reboot the Pi if you have made changes


```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
#  ADAFRUIT TEMPERATURE & HUMIDITY SENSOR
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

# Source: https://github.com/adafruit/Adafruit_CircuitPython_AHTx0/blob/main/examples/ahtx0_simpletest.py 

# SPDX-FileCopyrightText: 2021 ladyada for Adafruit Industries
# SPDX-License-Identifier: MIT

"""
Basic `AHTx0` example test
"""

import time
import board
import adafruit_ahtx0

# Create sensor object, communicating over the board's default I2C bus
i2c = board.I2C()  # uses board.SCL and board.SDA
# i2c = board.STEMMA_I2C()  # For using the built-in STEMMA QT connector on a microcontroller
sensor = adafruit_ahtx0.AHTx0(i2c)

while True:
    print("\nTemperature: %0.1f C" % sensor.temperature)
    print("Humidity: %0.1f %%" % sensor.relative_humidity)
    time.sleep(2)
```
