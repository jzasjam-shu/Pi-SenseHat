### Pins Reference
- **GND** → GND on Raspberry Pi
- **OUT** → GPIO4 (Pin 7) on Raspberry Pi
- **VCC** → 5V on Raspberry Pi

**Pins Reference**

Hint: Use command `pinout` in the Raspberry Pi terminal to view the GPIO pin information

**Tuning The Sensor**

The potentiometer (orange screw) on the right controls the sensitivity, and the potentiometer on the left controls the timeout (between 2.5 - 250 seconds). Here, both are turned fully anti-clockwise, meaning that the sensitivity and timeout are at their lowest. More information at ‘[Using a PIR sensor](https://projects.raspberrypi.org/en/projects/physical-computing/11)’.


### Prerequisite: Enable I2C for GPIO Pins Access

1. On the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, then select ‘Preferences’ and ‘Raspberry Pi Configuration’

<img width="400" alt="image" src="../images/config.png" />

2. In the Raspberry Pi Configuration window select the ‘Interfaces’ tab, then toggle I2C so that it is enabled and click ‘OK’

<img width="400" alt="image" src="../images/i2c.png" />

3. You will be required to reboot the Pi if you have made changes


```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# USE A PIR (PASSIVE INFRARED) SENSOR WITH RASPBERRY PI)
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

from gpiozero import MotionSensor
from datetime import datetime

pir = MotionSensor(4)

while True:
    pir.wait_for_motion()
    now = datetime.now()
    print("You Moved at ", now )
    pir.wait_for_no_motion()
    print("No Motion")
```
