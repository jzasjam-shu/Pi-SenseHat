### Wiring Instructions

**LED**
- Short Leg → GND rail/row on breadboard
- Long Leg → Same column as resistor
  
**Resistor**
- Resistor → Same column as LED long leg
- Resistor (other end) → Another column on the breadboard
  
**Button**
- In breadboard columns with pins aligned vertically
  
**Jumper Wires (M/M)**
- Black (in the example, use any colour)
- M → Same column as the button’s left legs 
- M → GND rail/row on the breadboard
  
**Jumper Wires (M/F)**
- Green (in the example, use any colour)
    - M → Same column as the resistor
    - F → GPIO20 (Pin 38) on Raspberry Pi
- Orange (in the example, use any colour)
    - M → Same column as the button’s right legs
    - F → GPIO21 (Pin 40) on Raspberry Pi
- Black (in the example, use any colour)
    - M → GND rail/row on the breadboard 
    - F → GND (Pin 6)

**Pins Reference**

Hint: Use command `pinout` in the Raspberry Pi terminal to view the GPIO pin information

```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# LIGHT AN LED USING A BUTTON  
# Modified from Source: https://gpiozero.readthedocs.io/en/stable/recipes.html#button-controlled-led
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

from gpiozero import LED, Button
from time import sleep

led = LED(20)           # LED with reference to GPIO Pin
button = Button(21)     # Button with reference to GPIO Pin

# Function for whan button is pressed
def btnDown():
    print("Button Down")

# Function for when buttion is released
def btnUp():
    print("Button Up")

# Create Event Handlers for functions
button.when_pressed = btnDown
button.when_released = btnUp

# Also, wait for button presses
while True:
    button.wait_for_press() # Wait until button is pressed
    print("Pressed!")
    led.toggle()        # Toggle the LED state
    sleep(0.5)
```
