### Pins Reference
- **GND** → GND on Raspberry Pi
- **VCC** → 3.3V or 5V on Raspberry Pi (via Resistor)
- **NC** → Not Connected
- **SIG** (Signal) → GPIO14 (Pin 8) on Raspberry Pi (via Resistor)

**Pins Reference**

Hint: Use command `pinout` in the Raspberry Pi terminal to view the GPIO pin information

### Enable I2C for GPIO Pins Access

1. On the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, then select ‘Preferences’ and ‘Raspberry Pi Configuration’

<img width="400" alt="image" src="../images/config.png" />

2. In the Raspberry Pi Configuration window select the ‘Interfaces’ tab, then toggle I2C so that it is enabled and click ‘OK’

<img width="400" alt="image" src="../images/i2c.png" />

3. You may be required to reboot the Pi if you have made changes


```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# USE THE SEEED STUDIO DHT11 TEMPERATURE & HUMIDITY SENSOR DHT11
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

# Source: https://pimylifeup.com/raspberry-pi-dht11-sensor

import time
import adafruit_dht
import board

dht_device = adafruit_dht.DHT11(board.D14) # Set the GPIO pin (eg. D4 = GPIO4)

while True:
    try:
        temperature_c = dht_device.temperature
        temperature_f = temperature_c * (9 / 5) + 32

        humidity = dht_device.humidity

        print("Temp:{:.1f} C / {:.1f} F    Humidity: {}%".format(temperature_c, temperature_f, humidity))
    except RuntimeError as err:
        print(err.args[0])

    time.sleep(2.0)
```
