# Raspberry Pi: Using GPIO Pins

---

## Temperature and Humidity Sensor (SEEED Studio: Grove DHT11)



### Pins Reference
- **GND** → GND on Raspberry Pi
- **VCC** → 3.3V or 5V on Raspberry Pi (via Resistor)
- **NC** → Not Connected
- **SIG** (Signal) → GPIO14 (Pin 8) on Raspberry Pi (via Resistor)

## Prerequisites

### Install Python Modules

1. Open a terminal window and use the commands below to install the required Python libraries (see the sensor documentation for more information).



To open a terminal on the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, select ‘Accessories’ then ‘Terminal’.

2. Run the following commands in the terminal window



NOTE: First `deactivate` any virtual environment
   
`sudo apt update && sudo apt upgrade -y`

`python3 -m pip install adafruit-circuitpython-dht`

`sudo pip3 install adafruit-circuitpython-ahtx0`

---

### Enable I2C for GPIO Pins Access

1. On the Raspberry Pi, click the ‘Raspberry Pi Menu’ button, then select ‘Preferences’ and ‘Raspberry Pi Configuration’



2. In the Raspberry Pi Configuration window select the ‘Interfaces’ tab, then toggle I2C so that it is enabled and click ‘OK’



3. You will be required to reboot the Pi if you have made changes


```python
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #
# USE THE SEEED STUDIO DHT11 TEMPERATURE & HUMIDITY SENSOR DHT11
# # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # # #

# Source: https://pimylifeup.com/raspberry-pi-dht11-sensor

import time
import adafruit_dht
import board

dht_device = adafruit_dht.DHT11(board.D14) # Set the GPIO pin (eg. D4 = GPIO4)

while True:
    try:
        temperature_c = dht_device.temperature
        temperature_f = temperature_c * (9 / 5) + 32

        humidity = dht_device.humidity

        print("Temp:{:.1f} C / {:.1f} F    Humidity: {}%".format(temperature_c, temperature_f, humidity))
    except RuntimeError as err:
        print(err.args[0])

    time.sleep(2.0)
```
