# Sense HAT: Detecting movement with the accelerometer

```python
from sense_hat import SenseHat
import time

sense = SenseHat()
```

```python
# Set pixel start position
x = 3
y = 3

# Clamp pixel position to 0-7
def clamp(value, min_value=0, max_value=7):
    return min(max_value, max(min_value, value))
```

```python
while True:

    # Read accelerometer
    accel = sense.get_accelerometer_raw()

    # Use tilt as velocity
    x += accel["x"] * 1
    y += accel["y"] * 1

    # Keep within display bounds
    x = clamp(x)
    y = clamp(y)

    # Draw pixel
    sense.clear()
    sense.set_pixel(int(round(x)), int(round(y)), 255, 255, 255)

    time.sleep(0.1)
```
